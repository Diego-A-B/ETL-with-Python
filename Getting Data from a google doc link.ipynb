```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to get the HTML content of the Google Doc
def fetch_google_doc(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.text
    else:
        raise Exception(f"Failed to fetch the page, status code: {response.status_code}")

# Function to extract table from the HTML content
def extract_table_from_html(html):
    soup = BeautifulSoup(html, 'html.parser')
    table = soup.find('table')  # Finds the first table in the document
    
    if not table:
        raise Exception("No table found in the document")
    
    # Check if table has headers
    headers = [header.get_text() for header in table.find_all('th')]
    
    # If no headers (th), use the first row as headers
    if not headers:
        first_row = table.find_all('tr')[0]
        headers = [cell.get_text() for cell in first_row.find_all('td')]
    
    rows = []

    # Extract data from each subsequent row
    for row in table.find_all('tr')[1:]:  # Start from second row if header is used
        cells = row.find_all('td')
        row_data = [cell.get_text() for cell in cells]
        
        # Ensure row_data matches the number of headers
        if len(row_data) == len(headers):
            rows.append(row_data)

    return headers, rows

# Function to convert table data into a pandas DataFrame
def convert_to_dataframe(headers, rows):
    df = pd.DataFrame(rows, columns=headers)
    return df

# Example Google Docs published link
#google_doc_url =

# Fetch the content and extract table data
html_content = fetch_google_doc(google_doc_url)
headers, rows = extract_table_from_html(html_content)

# Convert the extracted data into a DataFrame
df = convert_to_dataframe(headers, rows)
df['y-coordinate'] = df['y-coordinate'].astype(int)
df['x-coordinate'] = df['x-coordinate'].astype(int)


# Display the DataFrame
df

```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def fetch_table_from_google_doc(url):
    # Fetch the HTML content from the provided URL
    response = requests.get(url)
    
    if response.status_code != 200:
        raise Exception(f"Failed to fetch the page, status code: {response.status_code}")

    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find the table element in the HTML
    table = soup.find('table')
    
    if not table:
        raise Exception("No table found in the document.")
    
    # Extract headers from the table (if present)
    headers = [header.text.strip() for header in table.find_all('th')]
    
    # Extract rows and cell data
    rows = []
    for row in table.find_all('tr'):
        cells = [cell.text.strip() for cell in row.find_all('td')]
        if cells:
            rows.append(cells)
    
    # Create a DataFrame using the rows and headers
    df = pd.DataFrame(rows, columns=headers if headers else None)
    
    return df

# Example: Replace with your actual Google Docs URL
url = url
df = fetch_table_from_google_doc(url)

df
```
